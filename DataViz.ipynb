```python
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
with open("timeseries.json", "rt") as INFILE:
    data=json.load(INFILE)
```

```python
data
```

```python
datalist=data['data']

datalist
```

```python
dates=[dictionary['date'] for dictionary in datalist ]
dates.sort()
dates
```

```python
def parse_date(datestring):
    """ Convert a date string into a pandas datetime object """
    return pd.to_datetime(datestring, format="%Y-%m-%d")
```

```python
startdate=parse_date(dates[0])
enddate=parse_date(dates[-1])
print ( enddate - startdate
      )
```

```python
index=pd.date_range(startdate, enddate, freq='D')
timeseriesdf=pd.DataFrame(datalist)
timeseriesdf
```

```python
timeseriesdf['name'].unique()
```

```python
England_df = timeseriesdf[timeseriesdf['name'] == "England"]
```

```python
Northern_Ireland_df = timeseriesdf[timeseriesdf['name'] == "Northern Ireland"]
```

```python
Scotland_df = timeseriesdf[timeseriesdf['name'] == "Scotland"]
```

```python
Wales_df = timeseriesdf[timeseriesdf['name'] == "Wales"]
```

```python
England_df = England_df.drop('name', axis=1)
England_df.rename(columns={'newCases': 'England_NewCases'}, inplace=True)
England_df
```

```python
Northern_Ireland_df = Northern_Ireland_df.drop('name', axis=1)
Northern_Ireland_df.rename(columns={'newCases': 'Northern_Ireland_NewCases'}, inplace=True)
Northern_Ireland_df
```

```python
Scotland_df = Scotland_df.drop('name', axis=1)
Scotland_df.rename(columns={'newCases': 'Scotland_NewCases'}, inplace=True)
Scotland_df
```

```python
Wales_df = Wales_df.drop('name', axis=1)
Wales_df.rename(columns={'newCases': 'Wales_NewCases'}, inplace=True)
Wales_df
```

```python
England_df['date'] = England_df['date'].apply(parse_date)
England_sr =  pd.Series(England_df['England_NewCases'].tolist(), England_df['date'])
England_sr.asfreq('D')
England_sr = England_sr.reindex(index,fill_value=0.0)
England_sr.plot()
```

```python
Northern_Ireland_df['date'] = Northern_Ireland_df['date'].apply(parse_date)
Northern_Ireland_sr =  pd.Series(Northern_Ireland_df['Northern_Ireland_NewCases'].tolist(), Northern_Ireland_df['date'])
Northern_Ireland_sr.asfreq('D')
Northern_Ireland_sr = Northern_Ireland_sr.reindex(index,fill_value=0.0)
Northern_Ireland_sr.plot()
```

```python
Scotland_df['date'] = Scotland_df['date'].apply(parse_date)
Scotland_sr =  pd.Series(Scotland_df['Scotland_NewCases'].tolist(), Scotland_df['date'])
Scotland_sr.asfreq('D')
Scotland_sr = Scotland_sr.reindex(index,fill_value=0.0)

Scotland_sr.plot()
```

```python
Wales_df['date'] = Wales_df['date'].apply(parse_date)
Wales_sr =  pd.Series(Wales_df['Wales_NewCases'].tolist(), Wales_df['date'])
Wales_sr.asfreq('D')
Wales_sr = Wales_sr.reindex(index,fill_value=0.0)
Wales_sr.plot()
```

```python
final_data = pd.concat([England_sr,Northern_Ireland_sr,Scotland_sr,Wales_sr], axis=1)
final_data.set_axis(['England_New_cases', 'Northern_Ireland_New_cases', 'Scotland_New_cases', 'Wales_New_cases'], axis=1, inplace=True)
final_data
```

```python
final_data.dtypes
```

```python
final_data['Northern_Ireland_New_cases'] = final_data['Northern_Ireland_New_cases'].astype(int)
final_data['Scotland_New_cases'] = final_data['Scotland_New_cases'].astype(int)
final_data['Wales_New_cases'] = final_data['Wales_New_cases'].astype(int)
```
